```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_df = pd.read_parquet("artifacts/data/output/train.parquet")
train_df
```

```python
train_df.target.value_counts().plot(
    kind = "bar",
    title = "Distribution of credit cases by gender",
    xlabel = "gender",
    ylabel = "count",
    color = "grey"
)
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
train_df.marital_status.value_counts().plot(
    kind = "bar",
    title = "Distribution of credit cases by marital status",
    xlabel = "marital status",
    ylabel = "count",
)
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
train_df.income.plot(
    kind = "hist",
    title = "Distribution of credit cases by income",
    xlabel = "income",
    ylabel = "count",
    color = "grey"
)
```
